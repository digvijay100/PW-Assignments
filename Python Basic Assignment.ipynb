4. Discuss the different types of operators in Python and provide examples of how they are used

Ans : 1 Arithmetic Operators:
These operators perform basic arithmetic operations.

e.g. : + (addition), - (subtraction), * (multiplication), / (division), % (modulus), ** (exponent), // (floor division)


```python
a = 15
b = 10

print(a + b, end = " , ")
print(a - b, end = " , ")
print(a * b, end = " , ")
print(a / b, end = " , ")
print(a % b, end = " , ")
print(a ** b, end = " , ")
print(a // b)
```

  2. Comparison Operators:
These operators compare two values and return a boolean result (True or False).

e.g : == (equal to), != (not equal to), > (greater than), < (less than), >= (greater than or equal to), <= (less than or equal to)

```python
a = 15
b = 10

print(a == b, end = "  ,  ")
print(a < b, end = "  ,  ")
print(a > b, end = "  ,  ")
print(a != b, end = "  ,  ")
print(a <= b, end = "  ,  ")
print(a >= b)
```

 3 Logical Operators : Logical operators are used to combine conditional statements. e.g. and, or, not

```python
a = True
b = False

print(a and b , end ="  , ")
print(a or b , end = "  , ")
print(not a )
```

4 Bitwise Operators: Bitwise operators perform operations on binary representations of integers.

e.g.  & (AND), | (OR), ^ (XOR), ~ (NOT), << (left shift), >> (right shift)

```python
a = 5
b = 10

print(bin (a))
print(bin (b))

print ("a & b =>", bin(a & b) , " " , a & b )
print ("a | b =>", bin(a | b) , " ", a | b)
print ("a ^ b =>", bin(a ^ b) , " " , a & b )

print ("~a  =>", bin(~a) , " " , ~a )

print ("a << 2 =>", bin(a << 2) , " " , a << 2 )

print ("a >> 2 =>", bin(a >> 2) , " " , a >> 2 )

```

5 Assignment Operators: Assignment operators assign values to variables and perform operations in a single step.

e.g.: =, +=, -=, *=, /=, %=, **=, //=

```python
a = 10

a +=4

print(a)

a *=3
print(a)

a /= 2
print(a)
type(a)
```

5.   Explain the concept of type casting in Python with examples

Ans :  Process of converting one data type into another


```python
# Implicit typecasting : Python automatically converts one data type to another during an operation.
a = 5
b = 2.5

res = a + b

print(type(a), end = " , ")
print(type(b), end = " , ")

print(type(res), end = " , ")
```

```python
# Explicit typecasting : Programmer manually converts a data type using Python’s built-in functions, such as int(), float(), str(), and list()
ch = "5"

a = int(ch)

print(a , " ", type(a))

#del str
b = 20
data = str(b)

print(data , " " , type(data))

list1 = [1, 2, 3]

tup = tuple(list1)

print(tup , " " , type(tup))
```

7.  Describe the different types of loops in Python and their use cases with examples.

Ans :  for loop, while loop, nested loop


```python
# for loop : iterates over a sequence (like a list, tuple, dictionary, string, or range) and executes a code block for each item in the sequence.

for index in range(10):
  print(index , end = " , ")

Cities = ["Mumbai", "Pune", "Delhi"]

print("\n")

for city in Cities:
  print(city , end =" , ")


```

```python
#while loop : It repeats a code block as long as a specified condition is True. If the condition becomes False, the loop stops.

count = 0
while count < 10 :
  print(count, end =" ,")
  count += 1


```

```python
#nested loop : loop within another loop

for index1 in range(1, 6) :
  for index2 in range(1, 6) :
    print(index1 * index2, end = " ")
  print()
```
